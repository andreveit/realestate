```python



ap_dict = {
    'id':[],
    'created_at':[],
    'updated_at': [],
    'unit_types': [],
    'unit_flor': [],
    'area': [],
    'bedrooms': [],
    'bathrooms': [],
    'suites': [],
    'amenities':[],
    'listing_type':[],
    'parking_spaces':[],
    'description':[],
    'status':[],
    'zipcode':[],
    'level': [],
    'neighborhood':[],
    'complement':[],
    'price':[],
    'condo':[],
    'iptu':[],
    'total_price':[]    
}

for data in data_list:
    
    for ap in data['nearby']['search']['result']['listings']:
        ap = ap['listing']

        ap_dict['id'].append(ap['id'])
        ap_dict['area'].append(ap['usableAreas'][0] if len(ap['usableAreas'])>=1 else 0)
        ap_dict['created_at'].append(ap['createdAt'])
        ap_dict['updated_at'].append(ap['updatedAt'])
        ap_dict['unit_types'].append(ap['unitTypes'])
        ap_dict['unit_flor'].append(ap['unitFloor'])
        ap_dict['bedrooms'].append(ap['bedrooms'][0] if len(ap['bedrooms'])>=1 else 0)
        ap_dict['bathrooms'].append(ap['bathrooms'][0] if len(ap['bathrooms'])>=1 else 0)
        ap_dict['suites'].append(ap['suites'][0] if len(ap['suites'])>=1 else 0)
        ap_dict['amenities'].append(ap['amenities'])
        ap_dict['listing_type'].append(ap['listingType'])
        ap_dict['parking_spaces'].append(ap['parkingSpaces'])
        ap_dict['description'].append(ap['description'])
        ap_dict['status'].append(ap['status'])

        # Address
        ap_dict['zipcode'].append(ap['address']['zipCode'])
        ap_dict['level'].append(ap['address']['level'])
        ap_dict['neighborhood'].append(ap['address']['neighborhood'])
        ap_dict['complement'].append(ap['address']['complement'])


        # Pricing
        ap_dict['price'].append(ap['pricingInfos'][0]['price'])

        try:
            ap_dict['condo'].append(ap['pricingInfos'][0]['monthlyCondoFee'])
        except:
            ap_dict['condo'].append(np.nan)

        try:
            ap_dict['iptu'].append(ap['pricingInfos'][0]['yearlyIptu'])
        except:
            ap_dict['iptu'].append(np.nan)
        
        try:
            ap_dict['total_price'].append(ap['pricingInfos'][0]['rentalInfo']['monthlyRentalTotalPrice'])
        except:
            ap_dict['total_price'].append(np.nan)

df = pd.DataFrame(ap_dict)
df
```

```python
df.price = df.price.astype(int)
df.area = df.area.astype(int)
df['price_m2'] = df.price / df.area 
df.groupby('neighborhood').agg({'neighborhood':'count', 'price_m2':'mean', 'price':'mean', 'area':'mean'}).sort_values(by='price_m2', ascending=False).head(50)
```
